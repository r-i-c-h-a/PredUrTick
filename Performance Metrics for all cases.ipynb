```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import math
from sklearn.neighbors import NearestNeighbors
pd.options.mode.chained_assignment = None
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from statistics import mean 
```

```python
df = pd.read_csv("Dataset/ParkingData_Month_Time_Week.csv")
```

```python
data = df[['Latitude','Longitude','count','Day_Num','Issue Month','Time_Num','Street Name']]
```

```python
X = data[["Latitude","Longitude"]]
```

```python
min_pts_all = dict()
cluster_count_all = dict()
sill_score_all = dict()
knn_score_all = dict()
```

```python
for i in range(1,13):
    for j in range(0,7):
        for k in range(0,4):
            print(i,j,k)
            sample= data[(data['Day_Num']==j) & (data['Issue Month']==i) & (data['Time_Num']==k)]
            #print(sample.head(2))
            #print(sample.shape)
            sample_x = sample[['Latitude','Longitude']]
            
            # get min pts value -- heuristic method
            min_pts = round(math.log(sample.shape[0]))
            min_pts_all[(1,j,k)]=min_pts
            
            #build dbscan model for sample_x
            m = DBSCAN(eps=0.5, min_samples=min_pts)  #trial and error used to get eps value = 0.5
            m.fit(sample_x)
            
            clusters = m.labels_
            #print(set(clusters))
            
            #count number of clusters -1 means outlier
            n_clusters_ = len(set(clusters)) - (1 if -1 in clusters else 0)
            cluster_count_all[(i,j,k)]=n_clusters_
            
            #Sillhouette Score
            if(len(set(clusters))>1):
                sill_score_all[(i,j,k)]=metrics.silhouette_score(sample_x, clusters)
            
            clustered_x = sample_x
            clustered_x['cluster']=clusters
            clustered_x['street name']=sample['Street Name']
            #print(clustered_x.head())
            
            x = clustered_x[['Latitude','Longitude']]
            y = clustered_x['cluster']
            
            x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
            
            # Create KNN classifier
            knn = KNeighborsClassifier(n_neighbors = 3)
            # Fit the classifier to the data
            knn.fit(x_train,y_train)
            
            #KNN Score
            knn_score_all[(i,j,k)] = knn.score(x_test, y_test)
            
            
```

Mean Values for metric measures

Min Pts Values

```python
min_value = min(min_pts_all.values())
max_value = max(min_pts_all.values())
print("Range for min_pts values:", min_value, "to", max_value)
```

Number of clusters

```python
min_c_value = min(cluster_count_all.values())
max_c_value = max(cluster_count_all.values())
print("Range for number of clusters values:", min_c_value, "to", max_c_value)
```

Mean value for Sillhouette Score

```python
print("Mean Sillhouette Score:", mean(sill_score_all.values()))
```

Mean value for KNN Score

```python
print("Mean KNN Score:", mean(knn_score_all.values()))
```
