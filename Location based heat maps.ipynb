```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import folium
import requests
from folium.plugins import HeatMap
```

```python
#function to generate basic map of New York
def generateBaseMap(default_location=[40.693943, -73.985880], default_zoom_start=12):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
#generate heat map based on precinct
df_precinct = pd.read_csv("precinct.csv")
base_map = generateBaseMap()
HeatMap(data=df_precinct[['Latitude', 'Longitude', 'count']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist(), radius=10, max_zoom=13).add_to(base_map)
base_map.add_child(folium.ClickForMarker(popup='Potential Location'))
base_map.save("precinct map.html")
```

```python
#generate heat map based on street
df_street = pd.read_csv("street.csv")
print(len(df_street))
invalid = df_street.loc[(df_street['Latitude']=='n') & (df_street['Longitude']=='o')]
print(len(invalid))
```

```python
df_street = df_street[(df_street['Latitude']!='n') & (df_street['Longitude']!='o')]
len(df_street)
```

```python
#heat map for street data
base_map = generateBaseMap()
HeatMap(data=df_street[['Latitude', 'Longitude', 'count']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
base_map.add_child(folium.ClickForMarker(popup='Potential Location'))
```
