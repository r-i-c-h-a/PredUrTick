```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("Dataset/bounded box data.csv")
```

```python
df = df[pd.notnull(df['Vio_Time'])]
```

```python
df.shape
```

```python
df['Issue Date']=pd.to_datetime(df['Issue Date'])
```

```python
df['Day'] = df['Issue Date'].dt.day_name()
df['Day_Num'] = df['Issue Date'].dt.dayofweek
```

```python
df['Issue Month']=[i.month for i in df['Issue Date']]
```

```python
df['Vio_Time']=pd.to_datetime(df['Vio_Time'])
```

```python
df['Vio_Hour']=df.Vio_Time.dt.hour
```

```python
df.shape
```

```python
df['Time_Num']=[int(int(i)/6) for i in df['Vio_Hour']]
```

```python
df.to_csv('Dataset/ParkingData_Month_Time_Week.csv')
```
