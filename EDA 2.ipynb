```python
#plot for top 10 body types
body_type = list(df["Vehicle Body Type"].value_counts()[:10].keys())
body_type_10 = df.where(df["Vehicle Body Type"].isin(body_type)
```

```python
sns.countplot(body_type_10["Vehicle Body Type"])
```

```python
#plot for top 10 vehicle make
veh_make = list(df["Vehicle Make"].value_counts()[:10].keys())
veh_make_10 = df.where(df["Vehicle Make"].isin(veh_make))
```

```python
sns.countplot(veh_make_10["Vehicle Make"])
```

```python
# plot for top 10 violation codes
vio_code = list(df["Violation Code"].value_counts()[:10].keys())
vio_code_10 = df.where(df["Violation Code"].isin(vio_code))
```
