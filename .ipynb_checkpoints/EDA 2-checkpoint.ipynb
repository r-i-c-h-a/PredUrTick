```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("C:/Users/Mainaki Saraf/Desktop/Sem 5/DA/Project/Dataset/Sample_data_2017.csv")
```

```python
df.columns
```

```python
#plot for top 10 body types
body_type = list(df["Vehicle Body Type"].value_counts()[:10].keys())
body_type_10 = df.where(df["Vehicle Body Type"].isin(body_type))
```

```python
sns.countplot(body_type_10["Vehicle Body Type"]).set_title(" Bar chart for Top 10 Vehicle Body Type")
#Max for Suburban followed by 4-door sedan
```

```python
#plot for top 10 vehicle make
veh_make = list(df["Vehicle Make"].value_counts()[:10].keys())
veh_make_10 = df.where(df["Vehicle Make"].isin(veh_make))
```

```python
sns.countplot(veh_make_10["Vehicle Make"]).set_title("Bar chart for Top 10 Vehicle Make")
#Max for Ford followed by toyota and honda
```

```python
# plot for top 10 violation codes
vio_code = list(df["Violation Code"].value_counts()[:10].keys())
vio_code_10 = df.where(df["Violation Code"].isin(vio_code))
```

```python
sns.countplot(vio_code_10["Violation Code"]).set_title("Bar chart for Top 10 Violation Code")
#Max for violation No. 21 and No. 36
```

```python
#EDA for SUBN type cars
subn_df = df[df["Vehicle Body Type"]=="SUBN"]
#Plot for veh make for SUBN veh type
subn_veh_make = list(subn_df["Vehicle Make"].value_counts()[:10].keys())
subn_veh_make_10 = subn_df.where(subn_df["Vehicle Make"].isin(subn_veh_make))
sns.countplot(subn_veh_make_10["Vehicle Make"]).set_title("Bar chart for SUBN Car type's Top 10 Vehicle Make")
#Max is for Toyota and Honda
```

```python
#EDA for SUBN type toyota car colors
toy_subn_df = df[(df["Vehicle Body Type"]=="SUBN") & (df["Vehicle Make"]=="TOYOT")]
#Plot for car color of type SUBN and make Toyota
toy_subn_color = list(toy_subn_df["Vehicle Color"].value_counts()[:10].keys())
toy_subn_color_10 = toy_subn_df.where(toy_subn_df["Vehicle Color"].isin(toy_subn_color))
sns.countplot(toy_subn_color_10["Vehicle Color"]).set_title("Bar chart for SUBN Toyota Car's Top 10 Color")
#Maybe Grey? Requires cleaning of data
#Suburban Toyota Grey cars are more likely to be issued a parking ticket
```
