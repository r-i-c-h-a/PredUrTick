```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("C:/Users/Mainaki Saraf/Desktop/Sem 5/DA/Project/Dataset/Sample_data_2017.csv")
```

**Vehicle Body Types**

```python
#plot for top 10 body types
body_type = list(df["Vehicle Body Type"].value_counts().nlargest(10).keys())
body_type_10 = df.where(df["Vehicle Body Type"].isin(body_type))
```

```python
sns.countplot(body_type_10["Vehicle Body Type"]).set_title(" Bar chart for Top 10 Vehicle Body Type")
#Max for Suburban followed by 4-door sedan
```

Maximum tickets are issued to Suburban (SUBN) and 4-Door Sedan (4DSD) types

**Vehicle Make**

```python
#plot for top 10 vehicle make
veh_make = list(df["Vehicle Make"].value_counts().nlargest(10).keys())
veh_make_10 = df.where(df["Vehicle Make"].isin(veh_make))
```

```python
sns.countplot(veh_make_10["Vehicle Make"]).set_title("Bar chart for Top 10 Vehicle Make")
#Max for Ford followed by toyota and honda
```

Maximum tickets are issued to Ford and Toyota car makes

**Violation Codes**

```python
# plot for top 10 violation codes
vio_code = list(df["Violation Code"].value_counts().nlargest(10).keys())
vio_code_10 = df.where(df["Violation Code"].isin(vio_code))
```

```python
sns.countplot(vio_code_10["Violation Code"]).set_title("Bar chart for Top 10 Violation Code")
#Max for violation No. 21 and No. 36
```

Most tickets are issued under codes 21 and 36 <br>
21 - Parked where parking is not allowed by sign, street marking or traffic control device<br>
36 - Exceeding the posted speed limit in or near a designated school zone

**Suburban Car Type**

```python
#EDA for SUBN type cars
subn_df = df[df["Vehicle Body Type"]=="SUBN"]
#Plot for veh make for SUBN veh type
subn_veh_make = list(subn_df["Vehicle Make"].value_counts().nlargest(10).keys())
subn_veh_make_10 = subn_df.where(subn_df["Vehicle Make"].isin(subn_veh_make))
sns.countplot(subn_veh_make_10["Vehicle Make"]).set_title("Bar chart for SUBN Car type's Top 10 Vehicle Make")
#Max is for Toyota and Honda
```

```python
#EDA for SUBN type toyota car colors
toy_subn_df = df[(df["Vehicle Body Type"]=="SUBN") & (df["Vehicle Make"]=="TOYOT")]
#Plot for car color of type SUBN and make Toyota
toy_subn_color = list(toy_subn_df["Vehicle Color"].value_counts().nlargest(10).keys())
toy_subn_color_10 = toy_subn_df.where(toy_subn_df["Vehicle Color"].isin(toy_subn_color))
sns.countplot(toy_subn_color_10["Vehicle Color"]).set_title("Bar chart for SUBN Toyota Car's Top 10 Color")
#Maybe Grey? Requires cleaning of data
#Suburban Toyota Grey cars are more likely to be issued a parking ticket
```

Toyota Suburban Grey cars are more likely to be issued a parking ticket

**Ford Make**

```python
#EDA for Ford cars
ford_df = df[df["Vehicle Make"]=="FORD"]
#Plot for veh type for FORD
ford_veh_type = list(ford_df["Vehicle Body Type"].value_counts().nlargest(10).keys())
ford_veh_type_10 = ford_df.where(ford_df["Vehicle Body Type"].isin(ford_veh_type))
sns.countplot(ford_veh_type_10["Vehicle Body Type"]).set_title("Bar chart for FORD Cars Top 10 Vehicle Types")
#Max is for VAN followed by SUBN
```

Ford White Vans more likely to be issued parking tickets

**Violation Location**

```python
#Location of vehicle
#plot for top 10 body types
location = list(df["Violation Location"].value_counts().nlargest(10).keys())
location_10 = df.where(df["Violation Location"].isin(location))
loc_plot = sns.countplot(location_10["Violation Location"]).set_title("Bar chart for Top 10 Violation Locations")
plt.xticks(rotation=90)
#Max in 19th Precinct
```

Most (50%) tickets are issued by the 19th precinct that belongs to the Upper East Side area in Manhattan. On searching it was found that this is the most populated area of NYC.

**For 19th Precinct area**

```python
#street name for 19th Precinct
loc_19 =  df[df["Violation Location"]==19]
streets = list(loc_19["Street Name"].value_counts().nlargest(10).keys())
streets_10 = loc_19.where(loc_19["Street Name"].isin(streets))
sns.countplot(streets_10["Street Name"]).set_title("Bar chart for Top 10 streets in 19th Precinct")
plt.xticks(rotation=90)
#Max for 3rd Ave
```

3rd Avenue in the 19th Precinct is most suseptible to parking tickets. This street has many schools and supermarkets along its sides

```python
#violation code for 3rd Avenue
street_3ave =  df[(df["Violation Location"]==19) & (df["Street Name"]=="3rd Ave")]
vcode_3ave = list(street_3ave["Violation Code"].value_counts().nlargest(10).keys())
vcode_3ave_10 = street_3ave.where(loc_19["Violation Code"].isin(vcode_3ave))
sns.countplot(vcode_3ave_10["Violation Code"]).set_title("Bar chart for Top 10 Violations in 3rd Avenue,19th Precinct")
plt.xticks(rotation=90)
```

Most violations are 46,37 and 38

46 - Parking on the curb ie Double Parking<br>
37 - Parking in access of allowed time<br>
38 - Failing to show parking reciept<br>

These violations show that the cars stop along the curb and exceed the parking time limit. This must be because of the schools and supermarkets along the street where people don't park for long hours and hence look for parking near the supermarket or school to get back to their vehicle faster after buying grocery or dropping their children to school respectively.
